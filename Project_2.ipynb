# Part 1: Business Undestanding

<br>

Zillow is a real-estate marketplace company aimed at providing accurate data on homes throughout the United States to aid in the home buying and rental process. A key aspect of this data includes the estimated value of a given home, given a series of property features. 

To monitor the accuracy of their value predictions, Zillow tracks the log-error between their "Zestimate" and the actual sale price of a home. This is defined as:

$$
 logerror = log(Zestimate) - log(SalePrice)
$$

In an effort to improve their prediction algorithm, Zillow has tasked us with creating a model to evaluate their algorithm based on a set of property data. This data includes a series of property attributes as well as the log error of the Zestimate vs the Sale Price. 

As a secondary objective, Zillow has tasked us with aiding them with a new tool for classifying multi-use properties. To do this, they would like us to identify which properties have a large lot size vs home ratio, since these properties present potential for further construction.

Our goal is to construct a model that can accurately and accurately predict the effectiveness of Zillow's home value estimate on a home. Our secondary objective is to construct a model to classify the home/property ratio given a series of attributes. These model will not only help Zillow to better understand their home value prediction discrepancy, but should aid in the improvement of their prediction algorithms in the future.

```python
# Import the data 

import pandas as pd
import matplotlib.pyplot as plt

prop = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CONTENT/properties_2016.csv')
error = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/CONTENT/train_2016_v2.csv')
```

```python
# Change logerror value into a categorical variable in error dataframe

def error_check (row):
    if abs(row['logerror']) <= 0.025:
        return 'A'
    if abs(row['logerror']) <= 0.05:
        return 'B'
    if abs(row['logerror']) <= .1:
        return 'C'
    if abs(row['logerror']) <= 0.25:
        return 'D'
    return 'F'

error['Prediction Grade'] = error.apply(lambda row: error_check(row), axis = 1)

# Drop original logerror variable

error = error.drop('logerror', axis=1)
```

```python
z_data = pd.merge(prop, error, on = 'parcelid')

z_data.head()
```

## Evaluation Metrics

- **Accuracy:**
<br>
Ratio of correctly labeled predictions out of all predictions.
<br>
        
- **Precision:**
<br> 
Ratio of correctly labeled predictions out of all predictions for that classification. 
<br>
For example, the number of correctly labled A predictions out of all the A predictions.
<br>

- **Recall:**
<br> 
Ratio of correctly labeled predictions out of all actual members of that category
<br>
For example, the number of correctly labled Very Large properties out of all properties that actually had an a ratio of home to land greater than or equal to 6.
<br>
- **Specificity:**
<br>
The inverse of Recall: ratio of the number of predictions outside of a given category out of all actual members outside of the category.
<br>
For example, for the context of A grades: The number of correctly labeled B and C predictions (not A) that actually did not have an A prediction grade.
<br>
- **F-Measure:**
<br>
The F-Measure is the harmonic mean of precision and recall, with the highest score occuring when the two scores are balanced.
<br>
The F-Measure would be equal to  = $\frac{2 \cdot (Precision \cdot Recall)}{Precision + Recall}$
<br>
The F-Measure tends to be best if there exists a high class imbalance, since a high score takes into account both precision and recall.
- **ROC AUC Score:**
<br>
The ROC curve plots the true positive rate against the false positive rate at different probability classification thresholds.
<br>
The AUC (Area Under the Curve) measures the area below the ROC curve, which provides a measure of performance across all classification thresholds.
<br>
When performing multiclass predictions, multiple ROC curves are constructed, and the weighted average of the ROC AUC scores is calculated to return an Average ROC AUC.

## Train / Test Split 

For both the Prediction Grade classification model and the Home to Land Ratio classification model we want to split our data in such a way that allows for maximal train/test iterations while not being too computationally intensive. It is important that these models both perform at the highest level possible and that they be usable by the client for future data sets. In addition to these goals, we must ensure that both of these models have a suitable bias / variance trade off, where they can be properly generalized and applied to a larger population of data.

In order to achieve these objectives, we will utilize a cross validation technique. In this, we divide the data into a series of n folds, train the model on n-1 of the folds, and test on the final fold. This allows us to rearrange our data and repeat the process, testing on a different fold. In each iteration, we will utilize our evaluation metrics (with a focus on the ROC AUC score) to identify the overall performance of our model.

The best estimate of the overall output of our model would be to set n equal to the number of instances in our data set. However, we have over 60,000 rows, which means that we would need to perform a 60,000 fold cross validation, which would be incredibly computationally intensive. Exmpirical evidence suggests that there generally exists a good bias and variance tradeoff when we perform cross validation with 10 folds (n = 10). Therefore, we will move forward with 10 Fold Cross Validation for the Prediction Grade and Home to Land Ratio classification models.
